The researchers aim to collect clinical information to contribute to enhancing clinical outcomes by identifying endotypes and phenotypes that’ll help clinicians better comprehend pneumonia’s underlying mechanisms.

I want to first establish whether or not BMI affects this number, and in turn drives dire patient outcomes.

PREDICTION: I hope to determine whether the dataset can reveal how specific comorbidities, and/or specific combinations of comorbidities, affect the specific patient outcomes I mentioned earlier. Importing Packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

Importing dataset:

```python
cap_data = pd.read_csv('NACef_Data.csv')
```

```python
# quick sanity check, to make sure it's been imported properly
cap_data.head()
```

#### Data Prep + Manipulation

```python
# sanity check: calling specific columns being used for this specific analysis/investigation
cap_data[['record_id','age','bmi','comorbid','icu_death','gen_hosp_death','clinical_resp','secund_infec']].head()
```

```python
# creating new version of the data set
CAP_use = cap_data[['record_id','age','bmi','comorbid','icu_death','gen_hosp_death','clinical_resp','secund_infec']]
```

```python
# since there's multiple comorbidities as a list in each row, I'm extracting the unique values and then going to assign
# a number to each (still working out, since according to the dictionary paired with the data, there's more than 30 
# unique comorbidities.... it'll get confusing when there's an 11 vs a single 1 in the code, may use letters instead)

unique_comorb = CAP_use['comorbid'].str.split(',', expand=True)
# sorting alphabetically, to make later manipulations easier
unique_comorb[0:] = unique_comorb.apply(lambda x: pd.Series(sorted(x.dropna())), axis=1)
unique_comorb['record_id'] = CAP_use['record_id']
unique_comorb = unique_comorb.set_index('record_id').reset_index()
unique_comorb
```

```python
# counting comorbidities (based on if there's a value in the cell or not), 
# and creating a new column to track this for each patient(row)

#unfortunately, it seems like the code is adding an extra number (e.g 4 comorbid instead of 3 for the first row) 
# so this is a temporary fix
unique_comorb['#_of_comorbid'] = unique_comorb.count(axis="columns") - 1
unique_comorb
```

According to the CDC, the BMI categories are as follows:

<img src="CDC_obesity.png" width='500px'>

The following calculations will be done with only the rows that have BMI values. Thus, I have to first drop the rows that have no values. I will then create a column that labels the BMI values with the category, to further simplify the data for presentation.

- “underweight” = <18.5
- “moderate weight” = 18-25
- “overweight” = 25-30
- “obesity” = 30-35
- "moderate obesity" = 35-40
- “severe obesity” = >40


```python
# creating new DF with only rows that have BMI values, maintaining the original index and the patient IDs
BMI_calc = pd.DataFrame(CAP_use[['record_id','bmi']])
BMI_calc.dropna(subset=['bmi'], inplace=True)

#quick sanity check to make sure it dropped the rows properly
BMI_calc[0:8]
```

```python
BMI_calc.loc[(BMI_calc['bmi'] < 18.5),'BMI_status'] = 'underweight'
BMI_calc.loc[(BMI_calc['bmi'] > 18.5) & (BMI_calc['bmi'] < 25), 'BMI_status'] = 'moderate weight'
BMI_calc.loc[(BMI_calc['bmi'] > 25) & (BMI_calc['bmi'] < 30), 'BMI_status'] = 'overweight'
BMI_calc.loc[(BMI_calc['bmi'] > 30) & (BMI_calc['bmi'] < 35), 'BMI_status'] = 'obesity'
BMI_calc.loc[(BMI_calc['bmi'] > 35) & (BMI_calc['bmi'] < 40), 'BMI_status'] = 'moderate obesity'
BMI_calc.loc[(BMI_calc['bmi'] > 40), 'BMI_status'] = 'severe obesity'

#quick sanity check
BMI_calc[0:8]
```

I'll also be using dichotomous values for the "icu death", "general hospital death", and "secondary infection" (from hospital visit) columns. The values in the "secondary infection" column are currently nominal, so I'll start by reassigning to 0s if there's no value and 1s if there is — i.e. focusing on if there was/wasn't a secondary infection.


```python
CAP_use.loc[CAP_use['secund_infec'].notnull(), 'secund_infect_nom'] = 1
CAP_use.loc[CAP_use['secund_infec'].isnull(), 'secund_infect_nom'] = 0

#sanity check!
CAP_use.head()
```

In order to complete the comparisons between values, I'll create a new dataframe that joins the "unique_comorb" dataframe with "CAP_use" on "record_id" (so I can then plot them). 

```python
cu_plotdf = pd.merge(left=CAP_use, right=unique_comorb, on='record_id')

#quick sanity check
cu_plotdf
```

Now, I'll plot the number of comorbidities vs. each patient outcome as single points on three scatter plots. 

```python
outcomes = ['icu_death', 'gen_hospital_death', 'secund_infect_nom']

for outcome in outcomes:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='#_of_comorbid', y=outcome, data=cu_plotdf)
    plt.title(f'{outcome.replace("_", " ").title()} by Comorbidity Count')
    plt.show()
```

```python

plot_cs_counts = cu_plotdf.groupby(['#_of_comorbid', 'secund_infect_nom']).size().reset_index(name='counts')

plt.scatter(x = plot_cs_counts['#_of_comorbid'], y = plot_cs_counts['secund_infect_nom'], s=plot_cs_counts['counts'] * 3)
plt.xlabel('# of comorbidities')
plt.ylabel('secondary infections')
plt.title('Comorbidities vs. Secondary Infections')
```
