## basic functions

* write a function to add two numbers and add some basic comments
* run it once on two integers
* now run it once on an integer and a string
* can you handle this error better and return nan if one input ins a string? Use a general Exception

```python
def add_numbers(a, b):
    """
    a function to add two numbers
    input: food, integar; bar, integer
    output: integer
    
    """
    try:
        res = a*b
    except Exception as e:
        print('{} and {} needs to have the same type! {}'.format(a, b, e))
        res = np.nan
    return res
```

##### you can put "return" inside the "try" portion only;  better practice to only have one in entire function

```python
print(add_numbers(1,2))
```

```python
foo = "foo"
```

```python
print(add_numbers("test", "test"))
```

```python
print(add_numbers("test", 2))
```

## loops

* create a dictionary with 5 students names and their fav colors
* loop over the dictionary and print keys and values
* now choose any two students and only print names and collors for them
* now choose any one student and only print their fav color
* now choose any one student and do not print their fav color. Only print for everyone else
* can you write a while loop that loops over the keys and prints them?

```python
python_class = {
    'alejandro': 'purple',
    'himat': 'green',
    'alexander': 'blue',
    'carmen': 'yellow',
    'elizabeth': 'black'
}
```

```python
python_class
```

```python
for k, v in python_class.items():
    print(k,v)
```

```python
for k, v in python_class.items():
    if k == 'alejandro' or 'himat':
        print(k,v)
    else:
        print("not the chosen student")
```

```python
for k, v in python_class.items():
    if k == 'alejandro':
        print(k,v)
    elif k == 'himat':
        print(k,v)
    else:
        print("not the chosen student")
```

```python
for k, v in python_class.items():
    if k == 'alejandro':
        print(v)
        break
```

```python
for k, v in python_class.items():
    if k == 'alejandro':
        print(v)

```

```python
for k, v in python_class.items():
    if k == 'alejandro':
        continue
    print(v)
```

##### Note: "pass" used for python objects, a placeholder, while "continue" is if you know the purpose; i.e. they function the same, but use "continue"

```python
class_keys = list(python_class.keys())

idx=0
while idx < len(class_keys):
    print(class_keys[idx], python_class[class_keys[idx]])
    idx = idx+1
```
