### Extract DICOM metadata and analyze in pandas df
- extract relevant metadata and create a pandas dataframe
- create another pandas dataframe
- explore joins


```python
import os
```

```python
dicom_metadata1 = {
    'patient_id': ds.get('PatientID', np.nan),
    'filename': os.path.basename(filename),
    'modality': ds.get('Modality', np.nan),
    'mean_pixel_array': ds.pixel_array.mean()
}
```

```python
dicom_metadata1
```

```python
dicom_metadata1_df = pd.DataFrame(dicom_metadata1, index=['1CT1'])
```

```python
dicom_metadata1_df
```

### create another dicom metadata df for the same patient from a dict
- each key maps to a list of values
- use the same `patient_id` key
- create a diagnosis key with value "cancer"
- add an age key with value 56
- use same index
- name the dataframe `dicom_metadata2_df`

```python
dicom_metadata2_df = pd.DataFrame({
    'patient_id': ds.get('PatientID', np.nan),
    'diagnosis': 'cancer',
    'age': 56
}, index=['1CT1'])
```

## Joins
- simple join on 'patient_id'
- inner join

```python
pd.merge(
    left=dicom_metadata1_df,
    right=dicom_metadata2_df,
    on='patient_id',
    how='inner'
)
```

```python
# with indicator = True
pd.merge(
    left=dicom_metadata1_df,
    right=dicom_metadata2_df,
    on='patient_id',
    how='inner',
    indicator=True
)
```

### left join
- use keys from left frame only

```python
dicom_metadata1_df
```

### add a new row to `dicom_metadata1_df`
- index = 2CT1
- patient_id = '2CT1'
- filename = 'another_small_CT.dcm'
- modality = 'CT'
- mean_pixel_array = 100.2

```python
dicom_metadata1_df.loc['2CT1'] = {
    'patient_id': '2CT1',
    'filename': 'another_small_CT.dcm',
    'modality': 'CT',
    'mean_pixel_array': 100.2
}
dicom_metadata1_df
```

```python
dicom_metadata2_df
```

```python
pd.merge(
    left=dicom_metadata1_df,
    right=dicom_metadata2_df,
    on='patient_id',
    how='left',
    indicator=True
)
```

### right join
- use keys from right frame only

```python
dicom_metadata2_df
```

### add a new row to `dicom_metadata2_df`
- index = 3CT1
- patient_id = '3CT1'
- diagnosis = 'normal'
- age = 75

```python
dicom_metadata2_df.loc['3CT1'] = {
    'patient_id': '3CT1',
    'diagnosis': 'normal',
    'age': 75
}
```

```python
dicom_metadata2_df
```

```python
dicom_metadata1_df
```

```python
pd.merge(
    left=dicom_metadata1_df,
    right=dicom_metadata2_df,
    on='patient_id',
    how='right'
)
```

```python
pd.merge(
    left=dicom_metadata1_df,
    right=dicom_metadata2_df,
    on='patient_id',
    how='right',
    indicator=True
)
```

### outer join
- use keys from both frames

```python
pd.merge(
    left=dicom_metadata1_df,
    right=dicom_metadata2_df,
    on='patient_id',
    how='outer',
    indicator=True
)
```

## Plotting with seaborn
- https://seaborn.pydata.org/

```python
gene_tx_df = pd.read_csv('genes_transcripts.csv')
```

### scatterplot 
- `tx_length` for each `gene_name` colored by `biotype`

```python
sns.scatterplot(x='gene_name', y='tx_length', data=gene_tx_df, hue='biotype')
```

### integrate with matplotlib
- move legend

```python
fig, ax = plt.subplots()
ax = sns.scatterplot(x='gene_name', y='tx_length', data=gene_tx_df, hue='biotype')
ax.legend(bbox_to_anchor=(1,1))
plt.show()
```

### `sns.pairplot`
- quickly inspect data distributions by facets

```python
sns.pairplot(gene_tx_df, hue='gene_name')
```

### Facet grid
faceting histograms by subsets of data using displot
- visualize univariate or bivariate dist of data
- hist
- kde
- ecdf

```python
sns.displot(data=gene_tx_df, x='tx_length', 
            col='gene_name', height=4)
```

#### add a KDE curve
- kernel density estimate/PDF

```python
sns.displot(data=gene_tx_df, x='tx_length', 
            col='gene_name', height=4, kde=True)
```

### boxplots

```python
fig, ax = plt.subplots()
ax = sns.boxplot(x='gene_name', y='tx_length', hue='biotype', data=gene_tx_df)
ax.legend(bbox_to_anchor=(1,1))
plt.show()
```

### histplot

```python
fig, ax = plt.subplots()
ax = sns.histplot(data=gene_tx_df, x='tx_length', hue='gene_name')
plt.show()
```

### stripplots

```python
fig, ax = plt.subplots()
ax = sns.stripplot(x='gene_name', y='tx_length', hue='biotype', data=gene_tx_df)
ax.legend(bbox_to_anchor=(1, 1))
plt.show()
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.scatterplot(x='timepoint', y='signal', data=fmri, hue='region')
```

```python
#relplot for timepoints versus signal, colored by region and styled by event
ax = sns.relplot(x='timepoint', y='signal', hue='region', data=fmri, style='event')
plt.show()
```

```python
#boxplot of the same data
ax1 = sns.boxplot(x='timepoint', y='signal', hue='region', data=fmri)
plt.show()
```

###### **NOTE**: *there's a non-linear relationship between timepoint and signals for the different regions, and you can explore this using the lineplot function from seaborn!*

```python
sns.lineplot(x='timepoint', y='signal', hue='region', data=fmri)
plt.show()
```
