### Set operations

````python
first_genes_list = ['BRCA1', 'PTEN', 'PTEN', 'TMPRSS2', 'ERG']
````

````python
second_genes_list = ['BRCA2', 'PTEN', 'ERG', 'ERBB2']
````

### get intersect of genes in `first_genes_list` and `second_genes_list`

````python
set(first_genes_list) & set(second_genes_list)
````

### get list of genes only in `first_genes_list` and not in `second_genes_list`

````python
set(first_genes_list) - set(second_genes_list)
````

### Q: get list of genes only in `second_genes_list` and not in `first_genes_list`
- see example above and modify
- 1-2 mins

````python
set(second_genes_list) - set(first_genes_list)
````

### get list of genes in `first_genes_list` or `second_genes_list` or both
- union

````python
set(first_genes_list) | set(second_genes_list)
````

### get list of genes in `first_genes_list` or `second_genes_list`, but not both
- exclude intersect

````python
set(first_genes_list) ^ set(second_genes_list)
````

### Q: Can you get the same answer as above using a different approach learnt in the cells above?
- 1-2 mins

````python
(set(first_genes_list) | set(second_genes_list)) - (set(first_genes_list) & set(second_genes_list))
````

### set equality

````python
first_genes_list == second_genes_list
````

## Dictionaries
- set of key:value pairs
- keys have to be unique and immutable
- indexed by keys
- very powerful in python and commonly used data structure
- main operation is to store a key:value pair and look up the key

### Dictionary assignment and access
- `{}` assignment
- `dict()` constructor
- `dict` comprehension (next class)

````python
gene_location = {
    'TP53': 'chr17:7661779-7687538',
    'BRCA1' : 'chr17:43044295-43170245',
    'BRCA2': 'chr13:32315086-32400268',
}
````

````python
gene_location
````

````python
gene_location['TP53']
````

### Q: access `BRCA1` chromosomal location

````python
gene_location['BRCA1']
````

### You can only access dictionary keys by name
- will the following work?
- why or why not?

````python
gene_location[0]
````

````python
'BRCA1' in gene_location
````

````python
'RB1' not in gene_location
````

### update or modify a dictionary value
- replace `chr` with `chromosome`

````python
gene_location['BRCA1'] = gene_location['BRCA1'].replace('chr', 'chromosome')
````

````python
gene_location['BRCA1']
````

````python
gene_location
````
