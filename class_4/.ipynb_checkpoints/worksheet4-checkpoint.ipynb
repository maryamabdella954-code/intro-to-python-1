## Functions

#### Let's write a little function that can:
- take a parameter for a `file_name`
- read the file using pythons built-in functionality for files
- load the contents of the file into a dictionary
- return the dictionary

````python
def read_gene_desc_file(file_name):
    gene_desc = {}
    with open(file_name, 'r') as gene_desc_file:
        for line in gene_desc_file:
            stripped_line = line.strip()
            if '#' in stripped_line:
                continue
            gene_name, desc = stripped_line.split(',')
            gene_desc[gene_name] = desc
    return gene_desc
        
````

### Error and Exception handling
- First, the try clause (the statement(s) between the try and except keywords) is executed
- If no exception occurs, the except clause is skipped and execution of the try statement is finished
- If an exception occurs during execution of the try clause, the rest of the clause is skipped and exception executed

#### FileNotFoundError
- raise a specific error if the file is not found, with the path

````python
import os
def check_file_is_valid(file_name):
    try:
        f = open(file_name, 'r')
    except FileNotFoundError as e:
        print('{} not found at {}. Exception {}'.format(
            file_name, os.getcwd(), e
        ))
````

````python
check_file_is_valid('gene_description.csv')
````

#### More general exception

````python
def check_file_is_valid(file_name):
    try:
        f = open(file_name, 'r')
    except Exception as e:
        print('{} not found at {}. Exception {}'.format(
            file_name, os.getcwd(), e
        ))
````

````python
check_file_is_valid('gene_description.csv')
````

#### ValueError

````python
def return_gene_length_in_kb(gene_len_in_bp):
    try:
        kb_gene_len = int(gene_len_in_bp)/1000
        return kb_gene_len
    except ValueError as e:
        print('Invalid input "{}", enter numeric gene length in bp. Exception {}'.format(
            gene_len_in_bp, e
        ))
````

````python
return_gene_length_in_kb(10000)
````

````python
return_gene_length_in_kb('foo')
````

## An example function with decent comments
- Try to comment your code in this fashion for the collaborative exercise
- For single line code, you can comment using `#` right above the line of code

````python
def read_gene_desc_file(file_name):
    """
    This function reads a comma separated file,
    loads the contents into a dictionary and 
    returns the dictionary

    Args:
        file_name: The full path to the file
    Returns:
        gene_desc: a dictionary containing a gene name
        and a description
    """
    # init dict
    gene_desc = {}
    with open(file_name, 'r') as gene_desc_file:
        for line in gene_desc_file:
            stripped_line = line.strip()
            # skip header line
            if '#' in stripped_line:
                continue
            # split by comma and populate dict
            gene_name, desc = stripped_line.split(',')
            gene_desc[gene_name] = desc
    return gene_desc
````

````python
import requests

def read_ensembl_gene(gene_name):
    url = f'https://rest.ensembl.org/lookup/symbol/homo_sapiens/{gene_name}?content-type=application/json'
    link = requests.get(url)
    ens_gene_desc = link.json()
    return ens_gene_desc

read_ensembl_gene('TP53')
````

`Exercise 2`
- Construct a DNA sequence of length n=5 with your team
- Write a function to compute percentage of A bases
- Test on two other DNA sequences built with your team, including an empty string
- Raise an appropriate exception for an empty string


````python
dna_seq = ('A', 'G', 'T', 'A', 'C')

def percent_func(sequence, value):
    n = 0
    for v in range(len(sequence)):
        if sequence[v] == value:
            n = n +1
        else:
            continue
    try:
        percent = (n / len(sequence)) * 100
        return percent
    except ZeroDivisionError as e:
        print('Invalid input "{}", enter string with gene sequence. Exception {}'.format(
            sequence, e
        ))


percent_func(dna_seq, 'A')
````

`Exercise 3`
- Decide on a threshold voltage for hyperpolarization (or depolarizing) with your team
- Write a function that accepts a voltage and outputs whether hyperpolarizing, or depolarizing
- Test function on multiple voltage inputs, including non-numeric input
- Raise appropriate exceptions for invalid voltages  

````python

thresh_hyper_volt = -70

def volt_func(voltage):
    try:
        if voltage > thresh_hyper_volt:
            message = "The cell membrane is hyperpolarized"
        elif voltage < thresh_hyper_volt:
            message = "The cell membrane is in the process of depolarization"
        else:
            message = "The cell's membrane is currently at its resting potential (i.e. the threshold voltage for hyperpolarization)"
        return message
    except TypeError as e:
        print('Invalid input "{}", enter numeric value. Exception {}'.format(
            voltage, e
        ))
````
