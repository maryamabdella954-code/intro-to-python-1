NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

- Python library optimized for numerical computing
- ability to create and manipulate n-dimensional arrays, critical for ML/DL applications
- https://numpy.org/doc/stable/user/whatisnumpy.html#

Vectorized code
- `c = a * b`, where `a` and `b` each contain millions of elements or more
- pre-compiled C code

`ndarray` objects
- core of the package
- n dimensional array objects and operations on them

### Python lists versus Numpy arrays

Both are structures for storing and retrieving data

Python lists

- excellent, general-purpose containers
- “heterogeneous”, meaning that they can contain elements of a variety of types

What if we need to work on large quantities of homogenous data types in a very fast and efficient manner?

Numpy arrays

- All elements of the array must be of the same type of data
- Once created, the total size of the array can’t change
- The shape must be “rectangular”, not “jagged”; e.g., each row of a two-dimensional array must have the same number of columns

https://numpy.org/doc/stable/user/absolute_beginners.html

### Compare runtime behaviours between python lists and numpy arrays

- In the following code, we will calculate the sum of all squared numbers from 0 to 200 and see how much time the calculation will take. We do it 100 times and report the total time so that our measurement is accurate enough.
- Knowing how to optimize numpy takes great skill and practise

#### Numpy function

```python
def calculate_sum_np(end):
    nums = np.arange(end)
    return nums.dot(nums)
```

#### run it once

```python
calculate_sum_np(201)
```

#### timeit
- 100 times, 1 replicate each

```python
%timeit -n 100 -r 1 calculate_sum_np(201)
```

#### Python lists

```python
def calculate_sum_py_lists(end):
    res = []
    for i in range(end):
        res.append(i**2)
    return sum(res)
```

#### run it once

```python
calculate_sum_py_lists(201)
```

#### timeit

```python
%timeit -n 100 -r 1 calculate_sum_py_lists(201)
```

### Exercise 1

Implement a sigmoid function on toy dataset you create with your team
- The function should accept an input numpy array of length 10 and calculate the sigmoid function `f(z) = 1/(1+e^(-z))`
- This function only returns values between 0 and 1, and is used in models that return prob
- First, implement this function using numpy function for exponentiation
- Second, also include numpy functions for addition and division, in addition to exponentiation above
- Time the computations for the two cases listed above, using just one replicate and one loop. Which one runs the fastest?
- Now create a large input array of size 100000 random normal numbers, and pass it to the two functions. Which one runs the fastest?

```python
toy_data = np.array([4,6,5,8,10,3,13,4,2,9])
```

```python
def sig_func(array):
    neg = array*(-1)
    expo = np.exp(neg)
    sig = 1/(1+expo)
    return sig
```

```python
sig_func(toy_data)
```

```python
def sig_func2(array):
    neg = array*(-1)
    expo = np.exp(neg)
    sig = np.divide(1, (np.add(1,expo)))
    return sig
```

```python
sig_func2(toy_data)
```

```python
%timeit -n 1 -r 1 sig_func(toy_data)
```

```python
%timeit -n 1 -r 1 sig_func2(toy_data)
```

###### QA: it's clear that the first function runs faster, since more numpy functions increase the time

```python
large_arr = np.random.rand(100000)
```

```python
%timeit -n 1 -r 1 sig_func(large_arr)
```

```python
%timeit -n 1 -r 1 sig_func2(large_arr)
```

###### QA: The first function still seems to runs the fastest! (After running it multiple times, however, there are a few cases where the second is faster, but the majority of tries results in the first function being quicker by at least 1ms)

### Exercise 2

Implement a function that returns a MSE estimate, given two arrays as input

- The formula is available anywhere online, for eg. `https://statisticsbyjim.com/regression/mean-squared-error-mse/`
- Construct with your team, an input array of predictions, and an input array of observed, each with three values
- Calculate MSE using these two arrays as input
- How long does your code take to run, for one loop and one replicate?
- What if you passed two different sized input arrays: prediction array of size 1, and observed array of size 3? Does it work? Why, or why not?
- What if you now passed prediction array of size 2, and observed array of size 3? Does it work? Why, or why not?
- Modify your original function to raise an appropriate error or exception to handle unequal sized inputs, and run modified function on all cases

```python
predicted = np.array([1,2,3])
observed = np.array([4,5,6])
```

```python
def mse_func(array1, array2):
    ins_val = array2 - array1
    squ_val = ins_val**2
    mse_val = np.sum(squ_val) / array2.shape
    return mse_val
```

```python
mse_func(predicted, observed)
```

```python
%timeit -n 1 -r 1 mse_func(predicted, observed)
```

```python
new_pred = np.array([1])
new_obs = np.array([4,5,6])
```

```python
mse_func(new_pred, new_obs)
```

###### QA: This works, because the function will parse the equation over that single value in the array between all three values of the other array — i.e. (4-1)^2, (5-1)^2, etc. It's essentially treating the "1" as the prediction for all three observed values. 

```python
new_pred2 = np.array([1, 2])
new_obs2 = np.array([4,5,6])
```

```python
mse_func(new_pred2, new_obs2)
```

###### QA: This doesn't work, because the "1" and "2" values aren't assigned as predictions for a specific observed value, and the code will not parse the equation for both values with each of the second array's values — i.e. (4-1)^2, (4-2)^2, etc. The code uses the shape of the second vector to determine "n," the number of values, so having only one value in the first vector allows for reiteration with that single value over "n" without confusing which value pairs with which. (I'm assuming we would need matrix calculations if we were to compute a 3x1 and 2x1, but these are arrays not matrices)

```python
def new_mse_func(array1, array2):
    try:
        ins_val = array2 - array1
        squ_val = ins_val**2
        mse_val = np.sum(squ_val) / array2.shape
        return mse_val
    except ValueError as e:
        print('Exception {}. Make sure to input same vector shapes!'.format(e))
```

```python
new_mse_func(predicted, observed)
```

```python
new_mse_func(new_pred, new_obs)
```

```python
new_mse_func(new_pred2, new_obs2)
```
