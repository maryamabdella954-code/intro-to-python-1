# Series and DataFrame

- Series is a 1D array representing a single data column (i.e single column of an excel spreadsheet)
- A DataFrame is a 2D array used to represent tabular or spreadsheet-like data (i.e rows and columns)
- DataFrame can be thought of as a collection of Series (columns)
- Both Series and DataFrame have an index (an associate array of data labels)

## Exercise 1

- Decide on a threshold voltage for hyperpolarization (or depolarization) with your team
- Write a custom function that accepts a voltage and returns a `hyperpolarizing`  or `depolarizing` label as output for valid voltages. Raise appropriate exceptions for invalid voltages, including a `missing input` or `invalid input` label
 - Create a randomly simulated numpy array of voltages (n=10). Select two random positions in this array without replacement using `np.random.choice` and set the values to `np.nan` using numpy integer masks
- Convert numpy array to series
- Apply your custom function to the series and show the output
- Can you also check for missing values and print which rows of the series have those missing values?
- Can you create a new series where you drop the rows with missing values using `series.dropna()`?
- Can you create a new series where you fill the missing values with `0` using `series.fillna(0)`?
- https://numpy.org/doc/2.1/reference/random/generated/numpy.random.choice.html

````python
thresh_hyper_volt = -70

def volt_func(voltage):
    try:
        if voltage > thresh_hyper_volt:
            message = "hyperpolarizing"
        elif voltage < thresh_hyper_volt:
            message = "depolarizing"
        else:
            message = "Missing Input"
        return message
    except TypeError as e:
        print('Invalid input "{}", enter numeric value. Exception {}'.format(
            voltage, e
        ))
````

````python
np.random.seed(33)
random = np.random.randint(-100,200,10)
random_ints = np.array(random, dtype=float)
random_ints
````

````python
random_num = np.array(np.random.choice(random_ints, 2, replace=False))
random_num
````

````python
random_ints[np.isin(random_ints,random_num)] = np.nan
random_ints
````

````python
ran_ser = pd.Series(random_ints)
ran_ser
````

````python
ran_ser.apply(volt_func)
````

````python
ran_ser[ran_ser.isnull()]
````

````python
ran_ser2 = ran_ser.dropna()
ran_ser2
````

````python
ran_ser3 = ran_ser.fillna(0)
ran_ser3
````

- An example dataset could look like the following. Note that you will create two such datasets, with the same `electrode_name` in each, but different voltages

```
electrode_name, micro_volts
electrode1, 5
electrode2, 2
electrode3,  np.nan
electrode4, 10
```
- Load each dataset into a pandas series
- First, calculate the total voltage for each electrode using series addition. Explain output

````python
mv1 = [4, 5, np.nan, 13]
mv2 = [np.nan, 7, -3, 1]
electrodes = ['electrode1', 'electrode2', 'electrode3', 'electrode4' ]
````

````python
toy_set1 = pd.Series(mv1, electrodes)
toy_set1
````

````python
toy_set2 = pd.Series(mv2, electrodes)
toy_set2
````

````python
toy_set1 + toy_set2
````

###### Output: The first and third rows are NaN since you can't add a NaN value with an integar; meanwhile, the other rows simply added the microvolt values from the corresponding rows — i.e. row 1 + row 1, row 2 + row 2, etc.

````python
toy_set2.loc['electrode5'] = 3
toy_set2
````

````python
toy_set1+toy_set2
````

###### Output: Since there's no corresponding row in the first series to combine with the new row in the second, it returns NaN since the operation can't be performed.

## Exercise 3

- Randomly simulate 100 different EEG potentials with data labels and load into a pandas `series_A`
- Randomly select 50 rows of `series_A` using numpy and save in a new series called `ser_A_dup_data`. Modify the index labels of `ser_A_dup_data`
- Create a new `series_B` combining `series_A` with `ser_A_dup_data` (you just created a new series with duplicates). Use `pd.concat()` for this
- Calculate the 10%, 50% and 90% quantile on `series_B` using `series.quantile(q)` where q = [0.1, 0.5 or 0.9]
- Experiment with sort_index(), sort_values() and drop_duplicates() function on `series_B`. Run drop_duplicates() with no arguments, keep='first', keep='last' and keep=False. What do you observe?
- Experiment with `series_B.reset_index()` and `series_B.reset_index(drop=True)`. What do you observe? Explain your findings

````python
ran_pot = np.random.randint(-100,200,100)
ran_potentials = np.array(ran_pot, dtype=float)
series_A = pd.Series(ran_potentials)
series_A
````

````python
ser_A_dup_data = pd.Series(np.array(np.random.choice(ran_potentials, 50, replace=False)), dtype='float64')
ser_A_dup_data
````

````python
new_ser_a_index = [str(idx) + 'st EEG Potential' for idx in ser_A_dup_data.index]
ser_A_dup_data.index = new_ser_a_index
ser_A_dup_data
````

````python
series_B = pd.concat([series_A, ser_A_dup_data])
series_B
````

````python
series_B.quantile(0.1)
````

````python
series_B.quantile(0.5)
````

````python
series_B.quantile(0.9)
````

````python
series_B.sort_index()
````

````python
series_B.sort_values()
````

````python
series_B.drop_duplicates()
````

````python
series_B.drop_duplicates(keep='first')
````

````python
series_B.drop_duplicates(keep='last')
````

````python
series_B.drop_duplicates(keep=False)
````

###### Observation: The sort function only works with the values, since they're all integars and can thus be ordered, but the indices are a mix of strings and integars — there's no way to order them. The drop_duplicates function is fairly interesting, since it seems like the duplicate *values* are specifically dropped, and the "keep" argument determines which instance of the duplicate stays or if it stays at all (i.e. "False" deletes all duplicates, as indicated by the decreased length in that last experiment). 

````python
series_B.reset_index()
````

###### Findings/Observations: This is really interesting, since it appears to change the view of the series! It seems like a third dimension has been added, with the original index becoming another column, and the rows uniformly numbered. 

````python
series_B.reset_index(drop=True)
````

###### Findings/Observations: This performs a similar function as the version without an argument, but instead of preserving the original indices, it drops it completely and replaces them with uniform integars — the series remaines 2D.
