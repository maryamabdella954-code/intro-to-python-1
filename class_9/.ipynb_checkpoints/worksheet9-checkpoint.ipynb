```python
res = groupby_df.groupby('gene_type').agg(
    mean_gene_len_bp=('gene_len_bp','mean'),
    min_gene_len_bp=('gene_len_bp', 'min'),
    max_gene_len_bp=('gene_len_bp', 'max'),
    mean_time_0=('time.0','mean'),
    min_time_0=('time.0', 'min'),
    max_time_0=('time.0', 'max')
)
```

### pivot table
- alternative to groupby
- index = keys to group by on the index
- values = columns to aggregate

```python
groupby_df.groupby('gene_type')['gene_len_bp'].mean()
```

```python
#similar to the agg function, creating another DF
groupby_df.pivot_table(
    values='gene_len_bp', 
    index='gene_type',
    aggfunc='mean'
)
```

```python
groupby_df.groupby('gene_type')['time.0'].mean()
```

```python
groupby_df.pivot_table(
    values='time.0',
    index='gene_type',
    aggfunc='mean'
)
```

## Joins
- pandas merge function, SQL style joining operations
- https://pandas.pydata.org/docs/user_guide/merging.html

```python
d1 = {
    'gene_name': ['BRCA1', 'KRAS', 'HRAS', 'TP53', 'SMAD2'],
    'gene_length': [2, 4, 4, 6, 8]
}
left_df = pd.DataFrame(d1)
```

```python
left_df
```

```python
d2 = {
    'gene_name': ['BRCA1', 'KRAS', 'HRAS', 'TP53', 'MUC16'],
    'num_transcripts': [4, 6, 3, 7, 10]
}
right_df = pd.DataFrame(d2)
```

```python
right_df
```

#### simple join on 'gene_name'

```python
pd.merge(left=left_df, right=right_df, on='gene_name')
```

```python
pd.merge(
    left=left_df, 
    right=right_df, 
    on='gene_name', 
    indicator=True
)
```

### left join
- use keys from left frame only

```python
left_df
```

```python
pd.merge(
    left=left_df, 
    right=right_df, 
    on='gene_name', 
    how='left',
    indicator=True
)
```

### right join
- use keys from right frame only

```python
right_df
```

```python
pd.merge(
    left=left_df, 
    right=right_df, 
    on='gene_name', 
    how='right',
    indicator=True
)
```

### inner join
- intersection of keys from both frames

```python
pd.merge(
    left=left_df, 
    right=right_df, 
    on='gene_name', 
    how='inner',
    indicator=True
)
```

### outer join
- union of keys from both frames

```python
pd.merge(
    left=left_df, 
    right=right_df, 
    on='gene_name', 
    how='outer',
    indicator=True
)
```

```python
# Percentage of transcripts that are protein_coding for each gene
```

```python
# counts the number of occurences
df_counts = g_df.groupby(['gene_name', 'biotype']).size().reset_index(name='counts')
df_counts
```

```python
#calculates percentage of occurences, then only shows the protein_coding rows of those
df_counts['percentage'] = df_counts.groupby('gene_name')['counts'].transform(lambda x: (x / x.sum()) * 100)
df_counts[df_counts['biotype'] == 'protein_coding']
```

## Exercise 2
- You will create two pandas dataframes of your own, from your specific field, to demonstrate how joins work in pandas
  _________________
- Create two dataframes from two dictionaries, each with shape atleast (4, 3)
- Demonstrate inner join, left join, right join and outer join
- Make sure to include the appropriate data to demo all types of joins

```python
DLD1 = {
    'Paper Name' : ['Adlof et al 2021','Adlof et al 2021','Atl 2021','Archibald & Joanisse 2013'],
    'Diagnosis' : ['DLD','DLD+DYS','SLI','SLI'],
    'MeanAge' : ['95.75','95.75','58.2','100.8'],
    'Exp_Vocab' : ['EVT','EVT','PPVT_SS','NA'],
}
left_DF = pd.DataFrame(DLD1)
left_DF
```

```python
DLD2 = {
    'Paper Name' : ['Barak 2019','Busker 2011','Atl 2021','Archibald & Joanisse 2013'],
    'Diagnosis' : ['DLD','SLI','SLI','SLI'],
    'MeanAge' : ['67.17','69.83','58.2','100.8'],
    'Exp_Vocab' : ['Goralnik','NA','PPVT_SS','NA'],
}
right_DF = pd.DataFrame(DLD2)
right_DF
```

```python
# simple join, merging 
pd.merge(left=left_DF, right=right_DF, on='Paper Name')
```

```python
# left join
pd.merge(
    left=left_DF, 
    right=right_DF, 
    on='Paper Name', 
    how='left',
    indicator=True
)
```

```python
# right join
pd.merge(
    left=left_DF, 
    right=right_DF, 
    on='Paper Name', 
    how='right',
    indicator=True
)
```

```python
# inner join
pd.merge(
    left=left_DF, 
    right=right_DF, 
    on='Paper Name', 
    how='inner',
    indicator=True
)
```

```python
# outer join
pd.merge(
    left=left_DF, 
    right=right_DF, 
    on='Paper Name', 
    how='outer',
    indicator=True
)
```
